```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import string
from re import sub
import pandas as pd
import numpy as np
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
from unidecode import unidecode
plt.xticks(rotation=70)
pd.options.mode.chained_assignment = None
pd.set_option('display.max_colwidth', 100)
%matplotlib inline
```

```python
with open("/Users/debritbhattacharyya/Desktop/Suicide_Detection.csv", encoding='latin-1') as f:
    df = pd.read_csv(f)
f.close()
```

```python
df = df.dropna().drop_duplicates().reset_index(drop=True).rename(columns={'text':'comments'}).rename(columns={'class':'rate'})
```

```python
rws = df
```

```python
import contractions
```

```python
rws['no_contract'] = rws['comments'].apply(lambda x: [contractions.fix(word) for word in x.split()])
rws.head()
```

```python
rws['comments_str'] = [' '.join(map(str, l)) for l in rws['no_contract']]
rws.head()
```

```python
rws['tokenized'] = rws['comments_str'].apply(word_tokenize)
rws.head()
```

```python
rws['lower'] = rws['tokenized'].apply(lambda x: [word.lower() for word in x])
rws.head()
```

```python
punc = string.punctuation
rws['no_punc'] = rws['lower'].apply(lambda x: [word for word in x if word not in punc])
rws.head()
```

```python
stop_words = set(stopwords.words('english'))
rws['stopwords_removed'] = rws['no_punc'].apply(lambda x: [word for word in x if word not in stop_words])
rws.head()
```

```python
rws['pos_tags'] = rws['stopwords_removed'].apply(nltk.tag.pos_tag)
rws.head()
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
rws['wordnet_pos'] = rws['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
rws.head()
```

```python
wnl = WordNetLemmatizer()
rws['lemmatized'] = rws['wordnet_pos'].apply(lambda x: [wnl.lemmatize(word, tag) for word, tag in x])
rws.head()
```

```python
s=rws.pop('rate')
```

```python
n_rws=pd.concat([rws,s],1)
```

```python
rws=n_rws
rws
```

```python
file = pd.read_csv("Final_Training_Dataset.csv")
file_cleaned = file.dropna().drop_duplicates().reset_index(drop=True).rename(columns={'lemmatized':'title'})
file_cleaned
```

```python
def text_to_word_list(text, remove_bad_letters):
    ''' Pre process and convert texts to a list of words 
    method inspired by method from eliorc github repo: https://github.com/eliorc/Medium/blob/master/MaLSTM.ipynb'''
    text = remove_bad_letters(text)
    text = str(text)
    text = text.lower()

    # Clean the text
    text = sub(r"[^A-Za-z0-9^,!?.\/'+]", " ", text)
    text = sub(r"\+", " plus ", text)
    text = sub(r",", " ", text)
    text = sub(r"\.", " ", text)
    text = sub(r"!", " ! ", text)
    text = sub(r"\?", " ? ", text)
    text = sub(r"'", " ", text)
    text = sub(r":", " : ", text)
    text = sub(r"\s{2,}", " ", text)

    text = text.split()

    return text
```

```python
file_cleaned.title = file_cleaned.title.apply(lambda x: text_to_word_list(x, unidecode))
```
